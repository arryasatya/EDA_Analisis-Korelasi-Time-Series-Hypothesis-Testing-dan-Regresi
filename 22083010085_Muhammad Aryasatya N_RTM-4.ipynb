### a) Time Series Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
covid19 = pd.read_csv("covid_19_indonesia_time_series_all.csv", parse_dates=['Date'], index_col='Date')
covid19
```

##### Visualisasi data selama rentang beberapa tahun

```python
plt.figure(figsize=(10, 6))
plt.plot(covid19.index, covid19['New Cases'], marker='o', linestyle='-')
plt.title('Jumlah Kasus Baru Covid-19 Selama Rentang Beberapa Tahun')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus Baru')
plt.grid(True)
plt.show()
```

##### Pengelompokan (grouping) data time series berdasarkan bulan

```python
month_grouped = covid19.groupby(covid19.index.month).sum() 
month_grouped = month_grouped.rename_axis('Month')

month_grouped 
```

#### Pengelompokan (grouping) data time series berdasarkan hari

```python
month_grouped = covid19.groupby(covid19.index.day).sum() 
month_grouped 
```

##### Visualisasi bersama rata-rata mingguan kasus COVID-19 dan harian

```python
daily_grouped = covid19.resample('D').mean() 
week_grouped = covid19.resample('W').mean()

plt.figure(figsize=(12, 6))

plt.plot(daily_grouped.index, daily_grouped['New Cases'], label='Daily Cases')
plt.plot(week_grouped.index, week_grouped['  6New Cases'], label='Weekly Average')

plt.title('Daily and Weekly Average COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.grid()
plt.show()
```

### b) Hypotesis Testing

```python
from scipy import stats
import scipy.stats as st
import numpy as np
```

```python
from scipy.stats import ttest_1samp

rata_rata_harian = covid19.resample('D').mean()['New Cases'].mean()
statistic, p_value = ttest_1samp(covid19['New Cases'], popmean=rata_rata_harian)
print(f"Nilai p-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print(f"Hipotesis nol (H0) ditolak. Rata-rata lebih besar dari {rata_rata_harian}.")
else:
    print(f"Hipotesis nol (H0) diterima. Tidak cukup bukti untuk menyatakan rata-rata lebih besar dari {rata_rata_harian}.")
```

### C) Analisis Korelasi

```python
# Variabel X dan Y
X = covid19['Total Cases']
Y = covid19['Total Deaths']

# Menghapus missing value pada kolom Total Deaths
Y = Y.replace(0, np.nan).dropna()

# Menghitung nilai-nilai yang diperlukan
n = len(X)
sum_x, sum_y = X.sum(), Y.sum()
sum_x_square, sum_y_square = sum_x**2, sum_y**2
xi_square, yi_square = X**2, Y**2
sum_xi_square, sum_yi_square = xi_square.sum(), yi_square.sum()
xiyi = X * Y
sum_xiyi = xiyi.sum()

# Menghitung koefisien korelasi Pearson (r)
r = ((n * sum_xiyi) - (sum_x * sum_y)) / (np.sqrt((n * sum_xi_square - sum_x_square) * (n * sum_yi_square - sum_y_square)))
r
```

```python
X = covid19['Total Cases'].head(100)
Y = covid19['Total Deaths'].head(100)

Y = Y.replace(0, np.nan)
Y = Y.dropna()

n = 100

# X total
sum_x = X.sum()
sum_x_square = sum_x**2

# Xi^2
xi_square = X**2
sum_xi_square = xi_square.sum()

# Y total
sum_y = Y.sum()
sum_y_square = sum_y**2

# Yi^2
yi_square = Y**2
sum_yi_square = yi_square.sum()

# Xi * Yi
xiyi = X*Y
sum_xiyi = xiyi.sum()

p_kiri = (n*sum_xi_square) - sum_x_square
p_kanan = (n*sum_yi_square) - sum_y_square

# hitung koefisien korelasi
r = ((n*sum_xiyi) - (sum_x)*(sum_y))/((np.sqrt(p_kiri))*(np.sqrt(p_kanan)))
r
```

### d) Regresi dan evaluasi

```python
xbar = X.mean()
ybar = Y.mean()

# b1/slope
b1 = (sum_xiyi - (n*xbar*ybar))/(sum_xi_square - (n*np.sqrt(xbar)))

# b0/y-intercept
b0 = ybar - b1*xbar

# Model regresi
ytopi = b0 + b1*X
print('Model regresi = ',b0,' + ',b1,'*X')
```

```python
# Evaluasi model regresi
absolute = abs((Y-ytopi)/Y)

MAPE = (1/n) * (absolute.sum()) * 100/100
MAPE
```
